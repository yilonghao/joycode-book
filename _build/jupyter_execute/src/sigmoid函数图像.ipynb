```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def sigmoid(x):
    """sigmoid函数"""
    return 1 / (1 + np.exp(-x))
```

```python
x = np.linspace(-10, 10, 100)
y = sigmoid(x)
```

```python
plt.plot(x, y)
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('Sigmoid(x)')
plt.grid(True)
plt.show()
```
